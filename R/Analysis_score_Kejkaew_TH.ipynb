# Geting Rating data: combine files

ตัด 0001 ออกเนื่องจากมีไม่ครบทุก actions 

```python
import pandas as pd

avatar = ["Man", "Robot","Woman"]
sub_num = ['0002','0003','0004','0005','0006','0007','0008','0009','0010',
           '0011','0012','0013','0014','0015','0016','0017','0018','0019','0020','0021',
           '0022','0023','0024','0025','0026','0027','0028','0029','0030','0031','0032',
           '0033']

# columns=['Avatar','id', 'Animation','Rating','HRV_LFHF_mean']
all_data = pd.DataFrame()
```

```python
# sub_num = ['0002']
# avatar = ["Man"]
for i in range(len(sub_num)):
    for j in range(len(avatar)):
        filename = "MergeTh/MergedTh_" + sub_num[i] + "_" + avatar[j] + ".csv"
        print(filename)
        data = pd.read_csv(filename)
        data.loc[data.CharacterStatus.isna(), 'CharacterStatus'] = 'Walk'

        rating = data.loc[(data['User_Action'] == 'Answer'),['Form_Answer','CharacterStatus']] 
        print(rating.shape)
        group_df = data.groupby('CharacterStatus').agg({'HRV_LFHF': ['mean', 'std']})
        group_df.columns = ['HRV_LFHF_mean', 'HRV_LFHF_std']
        group_df = group_df.reset_index() 
        print(group_df.shape)
        merged_df = rating.merge(group_df, on='CharacterStatus', how='left')
        merged_df["sub_id"] = sub_num[i]
        merged_df["Avatar"] = avatar[j]
        print(merged_df.shape)
        all_data = pd.concat([all_data, merged_df], ignore_index=True)
```
