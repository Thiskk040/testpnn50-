```python
import pandas as pd

avatar = ["Man", "Robot","Woman"]
sub_num = ['0000','0001','0002','0003','0004','0005','0006','0007','0008','0009','0010',
           '0011','0012','0013','0014','0015','0016','0017','0018','0019','0020','0021',
           '0022','0023','0024','0025','0026','0027','0028','0029','0030']

# columns=['Avatar','id', 'Animation','Rating','HRV_LFHF_mean']
all_data = pd.DataFrame()
```

```python
# sub_num = ['0001']
# avatar = ["Man"]
for i in range(len(sub_num)):
    for j in range(len(avatar)):
        filename = "MergeFile/" + sub_num[i] + "_" + avatar[j] + ".csv"
        print(filename)
        data = pd.read_csv(filename)
        data.loc[data.Character_status.isna(), 'Character_status'] = 'Walk'

        rating = data.loc[(data['User_Action'] == 'Answer'),['Form_Answer','Character_status']].reset_index() 
        print(rating)
        # get only the question 1 - 8, but not question 0
        if rating.shape[0] == 8:
            rating = rating.drop(index=0)
        print(rating.shape)
        group_df = data.groupby('Character_status').agg({'HRV_LFHF': ['mean', 'std']})
        group_df.columns = ['HRV_LFHF_mean', 'HRV_LFHF_std']
        group_df = group_df.reset_index() 
        print(group_df.shape)
        merged_df = rating.merge(group_df, on='Character_status', how='left')
        merged_df["sub_id"] = int(sub_num[i]) +1
        merged_df["Avatar"] = avatar[j]
        print(merged_df.shape)
        all_data = pd.concat([all_data, merged_df], ignore_index=True)
```
